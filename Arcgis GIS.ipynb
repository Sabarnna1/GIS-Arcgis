```python
import arcgis
from arcgis.gis import GIS
from IPython.display import display
```

```python
gis = GIS('https://www.arcgis.com', 'username', 'password') ## ypu can create your own username and password from the given link to use the features.
```

```python
toolboxes = gis.content.search('travel', 'Geoprocessing Toolbox', 
                               outside_org=True, max_items=3)
```

```python
for toolbox in toolboxes:
    display(toolbox)
```
